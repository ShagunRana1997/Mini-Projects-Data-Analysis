```python
ledger = []

def add_transaction():
    date = input("Date: ")
    t_type = input("Type (Credit/Debit): ")
    amount = float(input("Amount: "))
    ledger.append({'date': date, 'type': t_type, 'amount': amount})

def view_ledger():
    balance = 0
    print("--- History ---")
    for record in ledger:
        print(record)
        if record['type'] == 'Credit':
            balance += record['amount']
        else:
            balance -= record['amount']
    print(f"Total Balance: {balance}")
```
