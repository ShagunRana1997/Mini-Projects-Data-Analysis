```python
library = [{'title': 'Python 101', 'author': 'Guido', 'status': 'Available'}]

def library_system():
    while True:
        action = input("\n1. Add 2. Issue 3. Return 4. Search 5. Exit: ")
        
        if action == '1':
            title = input("Title: ")
            author = input("Author: ")
            library.append({'title': title, 'author': author, 'status': 'Available'})
            
        elif action == '2':
            title = input("Title to issue: ")
            for book in library:
                if book['title'] == title and book['status'] == 'Available':
                    book['status'] = 'Issued'
                    print("Book issued.")
                    break
            else:
                print("Book not available or not found.")

        elif action == '3':
            title = input("Title to return: ")
            for book in library:
                if book['title'] == title:
                    book['status'] = 'Available'
                    print("Book returned.")

        elif action == '4':
            query = input("Search Title or Author: ")
            for book in library:
                if query in book['title'] or query in book['author']:
                    print(book)
                    
        elif action == '5':
            break
```
