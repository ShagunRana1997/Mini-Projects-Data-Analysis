```python
def atm_simulator():
    # Store user details in a dictionary
    user = {'user_id': {'pin': 1234, 'balance': 10000}}
    
    print("Welcome to the ATM!")
    entered_pin = int(input("Please enter your PIN: "))
    
    # Validate PIN
    if entered_pin != user['user_id']['pin']:
        print("Wrong PIN. Exiting.")
        return

    while True: # Keep looping until user exits
        print("\n1. Check Balance\n2. Deposit Cash\n3. Withdraw Cash\n4. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            print(f"Your balance is: {user['user_id']['balance']}")
        elif choice == '2':
            amount = float(input("Enter deposit amount: "))
            user['user_id']['balance'] += amount
            print("Deposit successful.")
        elif choice == '3':
            amount = float(input("Enter withdrawal amount: "))
            if amount <= user['user_id']['balance']:
                user['user_id']['balance'] -= amount
                print("Please take your cash.")
            else:
                print("Insufficient balance!")
        elif choice == '4':
            print("Thank you for using the ATM.")
            break
        else:
            print("Invalid option.")

# To run: atm_simulator()
```
