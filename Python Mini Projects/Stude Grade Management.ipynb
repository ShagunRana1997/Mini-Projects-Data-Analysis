```python
students = [] # List to store student dictionaries

def add_student():
    name = input("Enter Name: ")
    subject = input("Enter Subject: ")
    marks = float(input("Enter Marks: "))
    #Store record in a dictionary
    students.append({'name': name, 'subject': subject, 'marks': marks})

def show_stats():
    if not students:
        print("No data available.")
        return
    
    total_marks = sum(s['marks'] for s in students)
    average = total_marks / len(students)
    
    # Find topper
    highest_marks = -1
    topper_name = ""
    for s in students:
        if s['marks'] > highest_marks:
            highest_marks = s['marks']
            topper_name = s['name']
            
    print(f"Class Average: {average}")
    print(f"Topper: {topper_name} with {highest_marks} marks")

# Example usage:
# add_student()
# show_stats()
```
