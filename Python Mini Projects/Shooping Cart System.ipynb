```python
cart = []
menu_items = {'Apple': 50, 'Banana': 20, 'Milk': 30}

def shopping_cart():
    print("Available items:", menu_items)
    
    while True:
        action = input("Type 'add', 'remove', or 'done': ").lower()
        
        if action == 'done':
            break
        
        item = input("Enter item name: ")
        
        if action == 'add':
            if item in menu_items:
                cart.append(item)
                print(f"{item} added.")
            else:
                print("Item not in shop.")
                
        elif action == 'remove':
            if item in cart:
                cart.remove(item)
                print(f"{item} removed.")
            else:
                print("Item not in cart.")

    # Calculate total
    total = 0
    for item in cart:
        total += menu_items[item]
    
    print(f"Total Bill: ${total}")
```
