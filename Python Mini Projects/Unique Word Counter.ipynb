```python
def count_unique_words():
    text = input("Enter a paragraph: ")
    # Remove punctuation manually for beginner approach
    clean_text = text.replace('.', '').replace(',', '').lower() 
    words = clean_text.split()
    
    unique_words = set(words)
    print(f"Unique word count: {len(unique_words)}")
    
    # Frequency for top 5
    word_counts = {}
    for w in words:
        word_counts[w] = word_counts.get(w, 0) + 1
        
    #Sort by count (descending) and print top 5
    sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    print("Top 5 frequent words:", sorted_words[:5])
```
