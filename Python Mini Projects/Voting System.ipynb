```python
candidates = {'Alice': 0, 'Bob': 0}
voters_voted = []

def voting_system():
    while True:
        voter_id = input("Enter Voter ID (or 'stop'): ")
        if voter_id == 'stop':
            break
            
        if voter_id in voters_voted:
            print("You already voted!")
            continue
            
        print("Candidates: Alice, Bob")
        vote = input("Enter candidate name: ")
        
        if vote in candidates:
            candidates[vote] += 1
            voters_voted.append(voter_id)
            print("Vote cast.")
        else:
            print("Invalid candidate.")
            
    print("Results:", candidates)
    # Basic logic to find winner
    winner = max(candidates, key=candidates.get)
    print(f"Winner is: {winner}")
```
