```python
inventory = {}

def manage_inventory():
    while True:
        cmd = input("1. Add/Update 2. Calculate Value 3. Exit: ")
        
        if cmd == '1':
            name = input("Product Name: ")
            qty = int(input("Quantity: "))
            price = float(input("Price: "))
            inventory[name] = {'quantity': qty, 'price': price}
            
        elif cmd == '2':
            total_value = 0
            for prod, details in inventory.items():
                total_value += details['quantity'] * details['price']
            print(f"Total Inventory Value: ${total_value}")
            
        elif cmd == '3':
            break
```
