```python
contacts_list = []

def address_book_tuples():
    while True:
        cmd = input("1. Add 2. Search 3. Show All 4. Exit: ")
        
        if cmd == '1':
            name = input("Name: ")
            phone = input("Phone: ")
            contacts_list.append((name, phone))
            
        elif cmd == '2':
            search = input("Name to search: ")
            for contact in contacts_list:
                if contact[0] == search:
                    print(f"Found: {contact}")
                    
        elif cmd == '3':
            print(contacts_list)
            
        elif cmd == '4':
            break
```
