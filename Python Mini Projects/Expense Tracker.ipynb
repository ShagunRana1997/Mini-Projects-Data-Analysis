```python
expenses = []

def add_expense():
    cat = input("Category (Food/Travel): ")
    amt = float(input("Amount: "))
    date = input("Date: ")
    expenses.append({'category': cat, 'amount': amt, 'date': date})

def summary():
    totals = {}
    for e in expenses:
        cat = e['category']
        if cat in totals:
            totals[cat] += e['amount']
        else:
            totals[cat] = e['amount']
            
    print("Category Totals:", totals)
```
