```python
contacts = {}

def manage_contacts():
    while True:
        choice = input("\n1. Add/Update 2. Delete 3. Search 4. View All 5. Exit: ")
        
        if choice == '1':
            name = input("Name: ")
            phone = input("Phone: ")
            email = input("Email: ")
            contacts[name] = {'phone': phone, 'email': email}
            
        elif choice == '2':
            name = input("Name to delete: ")
            if name in contacts:
                del contacts[name]
                
        elif choice == '3':
            name = input("Search name: ")
            print(contacts.get(name, "Not found."))
            
        elif choice == '4':
            #sort alphabetically
            for name in sorted(contacts.keys()): 
                print(name, contacts[name])
                
        elif choice == '5':
            break
```
