```python
tasks = []

def todo_app():
    while True:
        cmd = input("1. Add 2. Mark Done 3. View 4. Exit: ")
        
        if cmd == '1':
            task_name = input("Task: ")
            tasks.append({'task': task_name, 'done': False})
            
        elif cmd == '2':
            idx = int(input("Enter task number to mark done (0 for first): "))
            if 0 <= idx < len(tasks):
                tasks[idx]['done'] = True
                
        elif cmd == '3':
            for i, t in enumerate(tasks):
                status = "Done" if t['done'] else "Pending"
                print(f"{i}. {t['task']} - [{status}]")
        
        elif cmd == '4':
            break
```
