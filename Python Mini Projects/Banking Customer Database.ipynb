```python
customers = {}

def banking_db():
    # ID is the key, value is another dictionary
    cid = input("Enter Customer ID: ")
    name = input("Name: ")
    c_type = input("Type (Savings/Current): ")
    bal = float(input("Initial Balance: "))
    
    customers[cid] = {'name': name, 'type': c_type, 'balance': bal}
    print("Customer added.")
    
    # Simple summary op
    print(f"Customer {name} has balance: {customers[cid]['balance']}")
```
