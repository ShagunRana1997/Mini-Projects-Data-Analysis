```python
questions = [
    {'q': 'Capital of France?', 'opts': 'a) Paris b) Rome', 'ans': 'a'},
    {'q': '2 + 2?', 'opts': 'a) 3 b) 4', 'ans': 'b'}
]

def play_quiz():
    score = 0
    for item in questions:
        print(item['q'])
        print(item['opts'])
        ans = input("Enter a or b: ")
        if ans == item['ans']:
            score += 1
    
    print(f"Your final score is: {score}/{len(questions)}")
```
