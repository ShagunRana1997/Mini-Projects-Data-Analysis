```python
# 5 rows, 5 seats per row. 'O' is open, 'X' is booked.
seats = [['O' for _ in range(5)] for _ in range(5)]

def book_ticket():
    for row in seats:
        print(row)
        
    r = int(input("Row (0-4): "))
    c = int(input("Seat (0-4): "))
    
    if seats[r][c] == 'O':
        seats[r][c] = 'X' # Mark as booked
        price = 100 if r < 2 else 80 # Simple pricing based on row
        print(f"Booked! Price: ${price}")
    else:
        print("Seat already taken.")
```
